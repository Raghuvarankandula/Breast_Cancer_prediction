```python
X =df.iloc[:,1:32].values
y =df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=22,stratify=y)
```

```python
## Because Support Vector Machine (SVM) optimization occurs by minimizing the decision vector w, 
##the optimal hyperplane is influenced by the scale of the input features 
##and it's therefore recommended that data be standardized (mean 0, var 1) prior to SVM model training.
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
model = SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cnf = confusion_matrix(y_test,y_pred)
sns.heatmap(cnf,annot=True,cmap='summer',fmt='g')
```

```python
acc = accuracy_score(y_test,y_pred)
print("Accuracy:",acc)
```

```python
print(classification_report(y_test,y_pred))
```
